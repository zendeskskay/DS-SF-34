## Part A | Model Fit and Customer Retention

```python
# TODO
train_df = pd.DataFrame({ 'Year' : [0, 1, 2, 3, 4, 5, 6, 7],
 'Retention_Rate' : [1, .869, .743, .653, .593, .551, .517, .491] })
train_df = train_df[['Year', 'Retention_Rate']]
train_df
```

```python
train_df.plot.scatter(x='Year',y='Retention_Rate')
```

```python
train_df['year_sqrt'] = train_df.Year.apply(lambda x: np.sqrt(x))
train_df.plot.scatter(x='year_sqrt',y='Retention_Rate')
```

```python
train_df['rate_sqrt'] = train_df.Retention_Rate.apply(lambda x: np.sqrt(x))
train_df.plot.scatter(x='Year',y='rate_sqrt')
```

```python
#with intercept
model = smf.ols(formula = 'np.reciprocal(Retention_Rate) ~ Year', data=train_df).fit()
model.summary()
```

The model with the intercept is way more predictive (R^2 = 0.982 vs R^2 = 0.468). It also has a much better p-value of 0.000153

```python
#without intercept
#X = df[ ['Year'] ] # X is a DataFrame
#y = df.Retention_Rate # y is a Series

#model2 = smf.OLS(y, X).fit()
#model2.summary()
```

```python
predict_X = pd.DataFrame({'Year' : [8,9,10,11,12]}, columns = ['Year'])
predict_X = sm.add_constant(predict_X)
```

```python
predict_y = model.predict(predict_X)
```

```python
predict_y
```

```python
test_df = predict_X
test_df['predicted_Retention_Rate'] = predict_y
test_df
```

```python
#test_df = test_df.drop('const', axis =1)
final = pd.concat([train_df,test_df])
final.drop(['const','rate_sqrt','year_sqrt'], axis = 1)
```

## Part B | One-Hot Encoding for Categorical Variables and SF Housing

```python
df = pd.read_csv(os.path.join('..', 'datasets', 'dataset-09-zillow.csv'), index_col = 'ID')
```

```python
df.drop(df[df.IsAStudio == 1].index, inplace = True)
```

```python
smf.ols(formula = 'SalePrice ~ Baths', data = df).fit().summary()
#in this case we are looking at bathrooms as if they were a ratio
#but intstead we want to look at it like an ordinal variable
```

> ### What's the bathrooms' distribution in the dataset?

```python
df.hist('Baths')
```

```python
df.plot.scatter(x='Baths',y='SalePrice')
```

> ### Let's keep properties with 1, 2, 3, or 4 bathrooms

```python
# TODO
#SalePrice = B0 + B1*Bath1 + B2*Bath2 + B3*Bath3 + B4*Bath4
# you will run 4 different models (each time eliminating one of the Bath variables) 
# ^ because you can deduce the value from the other values
#df =  df[(df.Baths >=1)&(df.Baths <=4)]
```

> ### Let's use `pandas`'s `get_dummies` to create a one-hot encoding of the `Bath` categorical feature

```python
# TODO
df = df[df.Baths.isin([1,2,3,4])]
df.Baths.value_counts(dropna = False).sort_index()
baths_df = pd.get_dummies(df.Baths, prefix = 'Bath')

#get_dummies is the same as writing out df['Bath_1'] = (df.Baths == 1)... and so on
```

```python
baths_df.rename(columns = {'Bath_1.0': 'Bath_1',
    'Bath_2.0': 'Bath_2',
    'Bath_3.0': 'Bath_3',
    'Bath_4.0': 'Bath_4'}, inplace = True)
```

```python
baths_df
```

```python
df = df.join([baths_df])
```

```python
df.columns
```

### Activity | Linear modeling using `Bath`'s one-hot encoding

```python
# TODO
model1 = smf.ols(formula = 'SalePrice ~ Bath_1 + Bath_3 + Bath_4', data=df).fit()
model1.summary()
```
